```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d
from tqdm import tqdm

plt.rcParams['figure.figsize'] = [20, 5]
```

```python
img = plt.imread('the-persistence-of-memory.jpg')
print(img.shape)
plt.imshow(img)
```

```python
img_bw = np.mean(img, axis=2)
plt.imshow(img_bw, cmap='gray')
```

```python
filter_x = np.array([[-1, 0, 1], 
                     [-2, 0, 2], 
                     [-1, 0, 1]])

filter_y = np.array([[-1, -2, -1], 
                     [ 0,  0,  0], 
                     [ 1,  2,  1]])

conv_x = np.abs(convolve2d(img_bw, filter_x, mode='same'))
conv_y = np.abs(convolve2d(img_bw, filter_y, mode='same'))

print(conv_x.shape)

fig, ax = plt.subplots(1, 2)
ax[0].imshow(conv_x, cmap='gray')
ax[1].imshow(conv_y, cmap='gray')
```

```python
edges = np.sqrt(conv_x ** 2 + conv_y ** 2)

plt.imshow(edges, cmap='gray')
```

```python
def find_energy_map(image):
    r, c = image.shape

    energy_map = np.zeros(image.shape)
    energy_map[-1] = image[-1]

    direction_map = np.zeros(image.shape, dtype=int)

    for i in reversed(range(r-1)):
        for j in range(c):
            left = max(0, j-1)
            right = min(c, j+2)
            direction = np.argmin(energy_map[i+1, left:right]) - 1
            if left == 0:
                direction += 1
            energy_map[i, j] = energy_map[i+1, j + direction] + image[i, j]
            direction_map[i, j] = direction
    
    return energy_map, direction_map

energy_map, direction_map = find_energy_map(edges)

plt.imshow(energy_map, cmap='gray')
```

```python
def find_seam_from(starting_point, dir_map):
    r, _ = direction_map.shape
    seam = np.zeros((r,), dtype=int)
    
    seam[0] = starting_point
    for i in range(1, r):
        seam[i] = seam[i-1] + dir_map[i-1, seam[i-1]]
        
    return seam

def find_seam(energy_map, direction_map):
    return find_seam_from(np.argmin(energy_map[0]), direction_map)

def show_seam(seam, img):
    r = img.shape[0]
    if img.ndim == 2:
        new_img = np.dstack((img, img, img))
        v_max = np.max(new_img, axis=(0, 1), keepdims=True)
        new_img = new_img / v_max
    else:
        new_img = img
    
    seam_color = np.array([1, 0, 0])
    
    for i in range(r):
        new_img[i, seam[i]] = seam_color
    
    return new_img

seam = find_seam(energy_map, direction_map)
seam_img = show_seam(seam, img_bw)

plt.imshow(seam_img)
```

```python
def remove_seam(seam, img):
    new_shape = list(img.shape)
    new_shape[1] -= 1
    new_img = np.zeros(new_shape, dtype=img.dtype)
    
    r = img.shape[0]
    c = img.shape[1]
    
    for i in range(r):
        mask = np.full((c,), True)
        mask[seam[i]] = False
        new_img[i] = img[i, mask]
    
    return new_img
    
new_img = remove_seam(seam, img)

plt.imshow(new_img, cmap='gray')
```

```python
def resize(img, steps):
    new_img = img_bw
    for i in tqdm(range(steps)):
        e_map, d_map = find_energy_map(new_img)
        current_seam = find_seam(e_map, d_map)
        new_img = remove_seam(current_seam, new_img)
    return new_img
    
plt.imshow(resize(img_bw, 150), cmap='gray')
```
